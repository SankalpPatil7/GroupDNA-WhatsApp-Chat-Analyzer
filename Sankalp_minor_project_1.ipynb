# Feature 1

## Reading the WhatsApp Chat File

```python
import numpy as np
from datetime import datetime
```

```python
FILE_NAME = "hostel_bois.txt"
with open(FILE_NAME,"r", encoding="utf-8") as file:
  lines = file.readlines()
print(f"Total lines in file: {len(lines)}")
# message = lines[0]
# print(message)
```

```python
messages =[]
system_messages = 0
media_messages = 0
deleted_messages = 0
```

```python
for line in lines:

    line = line.strip()

    if line == "":
        continue

    if " - " not in line:
        continue

    timestamp, content = line.split(" - ", 1)

    if ": " not in content:
        system_messages += 1
        continue

    sender, message = content.split(": ", 1)

    if "<Media omitted>" in message:
        media_messages += 1

    if "This message was deleted" in message:
        deleted_messages += 1

    date, time = timestamp.split(", ")

    dt = datetime.strptime(timestamp.strip(), "%d/%m/%y, %H:%M")

    messages.append({
        "datetime": dt,
        "date": date,
        "time": time,
        "hour": dt.hour,
        "sender": sender,
        "message": message
    })
```

```python
print("Parsed Messages:", len(messages))
print("System Messages:", system_messages)
print("Media Messages:", media_messages)
print("Deleted Messages:", deleted_messages)

print("\nFirst Parsed Message:")
print(messages[0])
```

# Feature 2 – Group Overview

```python
# -----------------------------
# GROUP OVERVIEW
# -----------------------------

total_messages = len(messages)

participants = set()

for msg in messages:
    participants.add(msg["sender"])

total_participants = len(participants)

start_date = messages[0]["date"]
end_date = messages[-1]["date"]

print("="*40)
print("          GROUP OVERVIEW")
print("="*40)

print(f"Total Messages     : {total_messages}")
print(f"Participants       : {total_participants}")
print(f"System Messages    : {system_messages}")
print(f"Media Messages     : {media_messages}")
print(f"Deleted Messages   : {deleted_messages}")
print(f"Date Range         : {start_date}  to  {end_date}")
```

# Feature 3 – Participant Analysis

```python
# -----------------------------
# PARTICIPANT ANALYSIS
# -----------------------------

message_count = {}

for msg in messages:

    sender = msg["sender"]

    if sender not in message_count:
        message_count[sender] = 1
    else:
        message_count[sender] += 1

print("="*40)
print("      PARTICIPANT ANALYSIS")
print("="*40)

for sender, count in message_count.items():
    print(f"{sender:<10} : {count}")

most_active = max(message_count, key=message_count.get)
least_active = min(message_count, key=message_count.get)

print("\nMost Active User :", most_active,
      "-", message_count[most_active], "messages")

print("Least Active User :", least_active,
      "-", message_count[least_active], "messages")
```

# Feature 4 – Activity Analysis

```python
# -----------------------------
# ACTIVITY ANALYSIS
# -----------------------------

hour_count = {}
day_count = {}

for msg in messages:

    hour = msg["hour"]
    date = msg["date"]

    # Count messages by hour
    if hour not in hour_count:
        hour_count[hour] = 1
    else:
        hour_count[hour] += 1

    # Count messages by date
    if date not in day_count:
        day_count[date] = 1
    else:
        day_count[date] += 1

# Find busiest hour and day
busiest_hour = max(hour_count, key=hour_count.get)
busiest_day = max(day_count, key=day_count.get)

print("=" * 45)
print("          ACTIVITY ANALYSIS")
print("=" * 45)

print("\nMessages by Hour")
print("-" * 25)

for hour in sorted(hour_count):
    print(f"{hour:02d}:00 - {hour_count[hour]}")

print("\nBusiest Hour :", f"{busiest_hour:02d}:00",
      "-", hour_count[busiest_hour], "messages")

print("\nBusiest Day :", busiest_day,
      "-", day_count[busiest_day], "messages")
```

# Feature 5 – Word Frequency Analysis

```python
# -----------------------------
# WORD FREQUENCY ANALYSIS
# -----------------------------

stop_words = {
    "the","is","a","an","to","of","and","in","on","for","at","with",
    "i","you","he","she","it","we","they",
    "hai","haan","ok","okay","yes","no","ya","bro","bhai"
}

word_count = {}

for msg in messages:

    text = msg["message"].lower()

    if "<media omitted>" in text:
        continue

    if "this message was deleted" in text:
        continue

    text = text.replace(".", "")
    text = text.replace(",", "")
    text = text.replace("!", "")
    text = text.replace("?", "")

    words = text.split()

    for word in words:

        if word in stop_words:
            continue

        if word not in word_count:
            word_count[word] = 1
        else:
            word_count[word] += 1

sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

print("="*45)
print("TOP 10 MOST USED WORDS")
print("="*45)

for word, count in sorted_words[:10]:
    print(f"{word:<15} {count}")
```

# Feature 6 – NumPy Activity Heatmap

```python
# -----------------------------
# NUMPY HEATMAP
# -----------------------------

participants = sorted(list(participants))

person_index = {}

for i, person in enumerate(participants):
    person_index[person] = i

heatmap = np.zeros((len(participants),24),dtype=int)

for msg in messages:

    row = person_index[msg["sender"]]
    col = msg["hour"]

    heatmap[row][col] += 1

print("="*70)
print("NUMPY ACTIVITY HEATMAP")
print("="*70)

print("Hours --->")
print("     ", end="")

for h in range(24):
    print(f"{h:>3}", end="")

print()

for i, person in enumerate(participants):

    print(f"{person:<6}", end=" ")

    for value in heatmap[i]:

        if value == 0:
            symbol = "."
        elif value <= 5:
            symbol = "░"
        elif value <= 15:
            symbol = "▒"
        else:
            symbol = "█"

        print(f"{symbol:>3}", end="")

    print()
```

# Feature 7 – Personality Detection

```python
# -----------------------------
# PERSONALITY DETECTION
# -----------------------------

print("="*50)
print("PERSONALITY ANALYSIS")
print("="*50)

for person in participants:

    total = 0
    night = 0
    long_msgs = 0

    for msg in messages:

        if msg["sender"] != person:
            continue

        total += 1

        if msg["hour"] >= 23 or msg["hour"] <= 4:
            night += 1

        if len(msg["message"].split()) >= 25:
            long_msgs += 1

    personality = "Regular Member"

    if total == min(message_count.values()):
        personality = "👻 Ghost"

    elif total == max(message_count.values()):
        personality = "💬 Spammer"

    elif night > total*0.5:
        personality = "🌙 Night Owl"

    elif long_msgs >= 5:
        personality = "📖 Story Teller"

    print(f"{person:<10} --> {personality}")
```
